##Create a function that checks whether a number is a prime number or not


```python
def isprime(num):
    if num > 1:
        for n in range (2, num):#for any number greater than 2
            if (num%n)==0: #if the remainder is zero:
                return False
        return True
    else:
        return False
print (isprime(89))  
print (isprime(78))                  
```

The fibonacci sequence is an infinite series of numbers, where each term is the sum of the two preceding ones

```python
def fib_seq(n):#define a function
    a=0
    b=1
    if n==0:
        print(a)
    else:
        print(a)
        print(b)    
    for i in range(2,n):
        c=a+b
        a=b
        b=c
        print(c)

fib_seq(15)#print the first 15 numbers in the fibonacci sequence
```
